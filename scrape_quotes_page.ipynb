# Scrape Quotes Page

Scrape and parse the quotes from this page of Seinfeld quotes.

## Setup

````python
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
````

````python
# Set variables for URL and header

# Seinfeld Quotes Page
url_quotes_page = "https://www.seinfeldscripts.com/seinfeld-quotes.html"

# Header
hdr = {'User-Agent': 'Mozilla/5.0 (iPad; CPU OS 12_2 like Mac OS X) AppleWebKit/605.1.15 (KHTML, like Gecko) Mobile/15E148'}
````

---

## Get Page Requests

````python
# 0. Use Request to add url and headers to request
req_quotes_page = Request(url_quotes_page, headers=hdr)
````

---

## Get Page HTML

````python
# 1. Get page HTML with urlopen
html_quotes_page = urlopen(req_quotes_page)
````

````python
# 2. Read HTML with bs 
bs_quotes_page = BeautifulSoup(html_quotes_page.read(), 'html.parser')
````

````python
# 3. Get page HTML 
bs_quotes_page
````

## Get Quotes

Quotes are stored on this page in the following format:

`<li><p>...</p></li>`

The quotes come in 2 formats:


```HTML
<li><p>Quote -Quotee</p></li>
```

and

```HTML
<li><p>Quotee 1: Quote <br/>
       Quotee 2: Quote</p></li>
```

````python
# First, find all list elements
list_elements = bs_quotes_page.find_all('li')
````

````python
# Next, find all list elements that contain paragraph tags
paragraph_elements = [list_element.find('p') for list_element in list_elements if list_element.find('p') is not None]

# print(len(paragraph_elements))
````

````python
multi_speaker_quotes = []

single_speaker_quotes = []
````

````python
for paragraph_element in paragraph_elements:
    # print(f'\n{paragraph_element.text}')
    if ':' in paragraph_element.text:
        multi_speaker_quotes.append(paragraph_element.text)
    else:
        single_speaker_quotes.append(paragraph_element.text)
````
