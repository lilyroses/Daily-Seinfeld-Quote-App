# Scrape Scripts Page

- URL: https://www.seinfeldscripts.com/seinfeld-scripts.html

Scrape and parse the script slink from this page for each Seinfeld episode.

---

## Setup

```python
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
```

```python
pwd
```

```python
# Set variables for URL and header

# Seinfeld Scripts Page
url_scripts_page = 'https://www.seinfeldscripts.com/seinfeld-scripts.html'

# Header
hdr = {'User-Agent': 'Mozilla/5.0 (iPad; CPU OS 12_2 like Mac OS X) AppleWebKit/605.1.15 (KHTML, like Gecko) Mobile/15E148'}
```

---

## Get Page Requests

```python
# 0. Use Request to add url and headers to request
req_scripts_page = Request(url_scripts_page, headers=hdr)
```

---

## Get Page HTML

```python
# 1. Get page HTML with urlopen
html_scripts_page = urlopen(req_scripts_page)
```

```python
# 2. Read HTML with bs 
bs_scripts_page = BeautifulSoup(html_scripts_page.read(), 'html.parser')
```

```python
# A BeautifulSoup(url, headers) object is of type
# bs4.BeautifulSoup
type(bs_scripts_page)

# A BeautifulSoup.bs.h1 object is of type
# bs4.element.Tag
type(bs_scripts_page.table)
```

```python
# 3. Get page HTML 

bs_scripts_page.findAll('td', {'align': 'middle'})
table_row_elements = bs_scripts_page.findAll('tr')

table_elements = bs_scripts_page.find_all('table')
```

```python
for table_element in table_elements:
    rows = [row for row in table_element.tr.next_siblings]
    for row in rows:
        print(row.text)
```
